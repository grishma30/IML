```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("/kaggle/input/nifty50-stock-market-data/ADANIPORTS.csv")
start_date = pd.to_datetime('2019-4-1')
end_date = pd.to_datetime('2019-09-30')                         
df['Date'] = pd.to_datetime(df['Date']) 
new_df = (df['Date']>= start_date) & (df['Date']<= end_date)
df2 = df.loc[new_df]
plt.figure(figsize=(10,10))
df2.plot(x='Date', y=['Open', 'Close']);
plt.suptitle('Opening/Closing stock prices \n 01-04-2019 to 30-09-2019', fontsize=12, color='black')
plt.xlabel("Date",fontsize=12, color='black')
plt.ylabel("$ price", fontsize=12, color='black')
plt.show()
```
