```python
#Grishma M. Jethva
#216270307051
#Write a Pandas program to create a plot of Open, High, Low, Close, 
#Adjusted Closing prices and Volume of given company between two 
#specific dates.

import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv('/kaggle/input/nifty50-stock-market-data/ADANIPORTS.csv')

start_date = '2008-11-27'  
end_date = '2009-11-27'    

df['Date'] = pd.to_datetime(df['Date'])  

filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

plt.figure(figsize=(12, 6))
plt.plot(filtered_df['Date'], filtered_df['Open'], label='Open Price', color='green')
plt.plot(filtered_df['Date'], filtered_df['Close'], label='Close Price', color='red')
plt.plot(filtered_df['Date'], filtered_df['High'], label='Close Price', color='blue')
plt.plot(filtered_df['Date'], filtered_df['Low'], label='Close Price', color='orange')



plt.title('Opening and Closing Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
