```python
#to find the maximum and minimum value of a given flattened array
#216270307051
#grishma m. jethva 

import numpy as np

a =  np. array([[1,2,3],[4,5,6]])
b=a.flatten()
print(b)
print("maximum of the given array")
print(np.max(b))
print("minimum of the given array")
print(np.min(b))
```

```python
#to compute the mean, standard deviation, and variance of a given 
#array along the second axis
#216270307051
#grishma m. jethva 

import numpy as np

a= np.arange(1,6)
print(a)
print("mean of the given array")
print(np.mean(a))
print("standard deviation")
print(np.std(a))
print("variance")
print(np.var(a))
```
