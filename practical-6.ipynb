```python
#Grishma M. jethva 
#216270307051
# to convert the first column of a DataFrame as a Series

import pandas as pd

information={"name":['xyz','abc','pqr'],"marks":[30,29,26]}

df=pd.DataFrame(data=information)
print("Info \n",df)

data1=df.iloc[0:,0]

print("\n First colum of data frame as a series \n\n",data1)
```

```python
#Grishma M. jethva 
#216270307051
# to create the mean and standard deviation of the data 
# of a given Series
import pandas as pd

a=[1,2,3,6,58,4]
se = pd.Series(a)
print("original series\n",se)

print("mean of above series",se.mean())
print("\n  standard deviation ",se.std())
```

```python
#Grishma M. jethva 
#216270307051
# to sort a given Series
import pandas as pd

color = ["red","white","yellow","pink"]
colors = pd.Series(color)
print("before sorting : \n\n",colors)
print("\nafter sorting:\n\n",colors.sort_values())
```
